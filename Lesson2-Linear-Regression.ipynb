# Lesson 2 - Linear Regression


Let there be a point (p,q) on a Cartesian plane and a line with equation 
$$ y =w_1x + w_2 $$
where the slope is $$ w_1 $$ and the y-intercept $$ w_2 $$


We can move the line closer to the point using 

## Absolute Trick

$$
f(x) = w_1x +p\alpha +w_2 +\alpha
$$

where the learning is $$ \alpha $$


## Square Trick

If the point is close to the line, we want to move the line at a smaller interval, but if the point is further away we want to move it by a larger interval. Square Trick takes the distance into the account.

$$
y=(w_1+p(q-q\prime)\alpha)x + (w_2 + (q-q\prime)\alpha)
$$
